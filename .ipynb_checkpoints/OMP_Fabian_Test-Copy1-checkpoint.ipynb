```python
import time
import math
import torch
import numpy as np
import matplotlib.pyplot as plt
import OMP_reconstruction
import pickle
import demo_heat_1d
import SQ_LASSO

dtype = torch.float
device_id = 'cuda'
```

```python
# Here we generate the data
M=32
D=11
N=5000

# sample D-dimensional array of random Tchebychev points in [-1,1]^D
samples = math.pi * (
    torch.rand(N, D, dtype=dtype, device=device_id, requires_grad=False) - 0.5
)
samples = samples.tan()
samples = samples / torch.sqrt(samples**2 + 1)

def hyp_cross(dim, R):
        if dim == 1:
            return [[k] for k in range(0, R + 1)]
        out = []
        for kk in hyp_cross(dim - 1, R):
            prod_val = 1
            for x in kk:
                prod_val *= max([1, abs(x)])
            rk = R // prod_val
            for r in range(0, rk + 1):
                out.append(kk + [r])
        return out

# create hyperbolic cross indices
indices = torch.round(
    torch.FloatTensor(hyp_cross(D, M)).to(dtype=dtype, device=device_id)
)
indices.requires_grad = False

values= demo_heat_1d.solve_heat_2(samples.cpu().numpy())
values = torch.tensor(values, dtype=dtype, device=device_id, requires_grad=False)[:,None]
values = values.contiguous()



# run OMP algorithm
start = time.time()
coeffs_rec, coeffs_lsr_rec, residuals= OMP_reconstruction.Tchebychev(samples,
                                                                     values,
                                                                     indices,
                                                                     num_iters=1000,
                                                                     lstsq_rec=True,
                                                                    )

# some evaluation metrics (OMP)
print("Time to compute OMP solution is", time.time() - start, "seconds")
print("Final residual of OMP solution is", residuals[0].item())
print(
    "Number of nonzero entries of OMP solution (with tolerance 1e-4):",
    torch.sum(torch.abs(coeffs_rec) > 1e-4).item(),
)

# evaluation of some metrics (Least Squares)
print("Residual of least squares solution is", residuals[1].item())


# Plot least squares coefficients
fig = plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.plot(coeffs_lsr_rec.cpu(), label="Reconstructed Coefficients")
plt.legend()
plt.title("Reconstructed Tchebychev Coefficients (Least Squares)")

# Plot OMP vs ground truth
fig = plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 1)
plt.plot(coeffs_rec.cpu(), label="Coefficients")
plt.legend()
plt.title("Computed Tchebychev Coefficients (OMP)")
```
